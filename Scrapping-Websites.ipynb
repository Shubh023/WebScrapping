```python
import pandas as pd
import numpy as np
import requests
import re
from bs4 import BeautifulSoup as bs
from lxml import html

page = requests.get('http://www.lycee-benjamin-franklin.fr/php5/spip/spip.php?article130')
tree = html.fromstring(page.content)
  
emails = tree.xpath('//span//strong/text()')
emails = [" ".join(emails[i].split()) for i in range(len(emails))]
#print(emails)
        
data = pd.DataFrame({'Emails':emails})
data
```
